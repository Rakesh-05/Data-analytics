```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

# Load the sales data with a different encoding
data = pd.read_csv("sales_data_sample.csv", encoding='latin1')
```

```python
# Display the first few rows of the data
print(data.head())
```

```python
data.head()
```

```python
# Display the data types and missing values
print(data.info())
print(data.isnull().sum())
```

```python
# Fill or drop missing values if any
data.fillna(0, inplace=True)
print(data.info())

# Add a 'Sales' column (Quantity * Price)
data['Sales'] = data['QUANTITYORDERED'] * data['PRICEEACH']
```

```python
# Summary statistics
print(data.describe())
```

```python
# Identify peak sales periods 
data['month'] = data['MONTH_ID']
monthly_sales = data.groupby('month')['Sales'].sum()
plt.figure(figsize=(12,6))
monthly_sales.plot(kind='bar')
plt.title('Sales of the Months')
plt.xlabel('Month of the Year')
plt.ylabel('Sales')
plt.show()
```

```python
# Store location analysis
store_location_sales = data.groupby('COUNTRY')['Sales'].sum()

plt.figure(figsize=(12,6))
sns.barplot(x=store_location_sales.index, y=store_location_sales.values)
plt.title('Sales by Country Location')
plt.xlabel('Stores Location')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```
